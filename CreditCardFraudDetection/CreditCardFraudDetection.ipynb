```python
fig, ax = plt.subplots(1, 2, figsize=(18,4))

amount_val = df['Amount'].values
time_val = df['Time'].values

sns.histplot(df['Amount'], ax=ax[0], kde=True, color='r')
ax[0].set_title('Распределение сумм транзакции', fontsize=14)
ax[0].set_xlim([min(df['Amount']), 150])

sns.histplot(df['Time'], ax=ax[1], kde=True, color='b')
ax[1].set_title('Распределение времени транзакции', fontsize=14)
ax[1].set_xlim([min(df['Time']), max(df['Time'])])

plt.show()
```

```python
from sklearn.preprocessing import RobustScaler
rs_scaler = RobustScaler()

df['scaled_amount'] = rs_scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df['scaled_time'] = rs_scaler.fit_transform(df['Time'].values.reshape(-1,1))

df.drop(['Time','Amount'], axis=1, inplace=True)
df.head()
```

```python
corr = df.corr()
plt.figure(figsize = (15,10))
sns.heatmap(corr, cmap='coolwarm_r')
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['Class'], axis=1)
y = df['Class']
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
import xgboost as xgb
model = xgb.XGBClassifier(n_estimators = 500, max_depth = 30, learning_rate = 0.01)
```

```python
model.fit(x_train, y_train.ravel())
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, precision_recall_curve, classification_report, precision_score, recall_score, accuracy_score
cnf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu", fmt='g')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')

print(classification_report(y_test, y_pred))
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=2)
x_train_s, y_train_s = sm.fit_sample(x_train, y_train.ravel())
```

```python
model.fit(x_train_s, y_train_s)
y_pred = model.predict(x_test)
```

```python
cnf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu", fmt='g')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')

print(classification_report(y_test, y_pred))
```
